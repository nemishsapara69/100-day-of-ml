# Matplotlib tutorial

```python
import matplotlib.pyplot as plt
%matplotlib inline## if u use this so dont write plt.show().
import numpy as np
```

```python
# simple exapmle
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
#plotting using matplotlib
## plt scatter

plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2d')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
#plt plot
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2d')
plt.show()
```

```python
## Creating subplots

plt.subplot(2,2,1) #2-nrows,2-ncol,1-position of graph or number
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(y,x,'g')
plt.subplot(2,2,3)
plt.plot(x,x,'b')
plt.subplot(2,2,4)
plt.plot(y,y,'y')
plt.show()
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y,label='2x')
plt.legend()

plt.show()
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")

plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)

plt.plot(x,y_sin,'r--')
plt.title('sine')
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cos')
plt.show()
```

```python
# Bar plot
x=[2,3,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
# histrogram
a=np.array([22,87,5,43,23,56,47,89,34,32,54,65,76,54])
plt.hist(a,bins=20)
plt.title('histrogram')
plt.show()
```

```python
#box plot using matplotlib

data=[np.random.normal(0,std,100) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
# pie chart
labels='Python','c++','java','R'
sizes=[215,345,432,124]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0)#explode 1st slice

#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
